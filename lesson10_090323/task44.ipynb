```python
import pandas as pd
import random
lst = ['robot', 'human', 'cat', 'dog', 'apple'] * 10  # можно писать сколько угодно наименований для увеличения количества столбцов
data = pd.DataFrame({'WhoAmI':lst})

unic = data[list(data)[0]].unique()
mlst = []
for i in data.index:
    temp = []
    for val in unic:
        temp.append(1 if data[list(data)[0]][i] == val else 0) 
    mlst.append(temp)    
    
data1 = pd.DataFrame(mlst, columns = unic)
data1.head(10)
```

Для частных случаев:

```python
import random
lst = ['robot', 'human', 'cat'] * 10  # для увеличения количества наименований столбцов добавлять (например): 'dog'
random.shuffle(lst)
data = pd.DataFrame({'WhoAmI':lst})
data.head(10)

# для каждого нового наименования столбца необходимо будет добовлять следующие 2 строки (например):
# data.loc[data['WhoAmI'] == 'dog', 'dog'] = '1'
# data.loc[data['WhoAmI'] != 'dog', 'dog'] = '0'
data.loc[data['WhoAmI'] == 'human', 'human'] = '1'
data.loc[data['WhoAmI'] != 'human', 'human'] = '0'
data.loc[data['WhoAmI'] == 'robot', 'robot'] = '1'
data.loc[data['WhoAmI'] != 'robot', 'robot'] = '0'
data.loc[data['WhoAmI'] == 'cat', 'cat'] = '1'
data.loc[data['WhoAmI'] != 'cat', 'cat'] = '0'

data.head(10)
```
